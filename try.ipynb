```python
def clean_col(df, col_name):
    df = df.copy()
    # df[col_name] = df[col_name].str.strip()
    df[col_name] = df[col_name].str.lower()
    df[col_name] = df[col_name].str.replace('_', ' ')
    df[col_name] = df[col_name].str.replace('-', ' ')
    # df[col_name] = df[col_name].str.translate(
    #     str.maketrans("", "", "!\"#$%&'()*+,-./:;<=>?@[\\]^_`{|}~")
    # )
    # df[col_name] = df[col_name].str.strip()

    return df
```

```python
df["comuna_ciudadano"] = df["comuna_ciudadano"].astype(int)

cols_to_clean = ['sexo', 'tipo_de_emprendimiento', 'barrio',
                 'línea_credito', 'idea_negocio']

for col in cols_to_clean:
    df = clean_col(df=df, col_name=col)

df
```

```python
df['monto_del_credito'] = df['monto_del_credito'].str.strip('$')
df['monto_del_credito'] = df['monto_del_credito'].str.replace(',', '')
df['monto_del_credito'] = df['monto_del_credito'].str.replace(' ', '')
df["monto_del_credito"] = df["monto_del_credito"].astype(float)
df["monto_del_credito"] = df["monto_del_credito"].astype(int)
df
```

```python
def correccion(fecha):
    componentes = fecha.split('/')
    if len(componentes[0]) == 4:
        nueva_fecha = '/'.join(reversed(componentes))
    else:
        nueva_fecha = fecha
    return nueva_fecha
    
df["fecha_de_beneficio"] = df["fecha_de_beneficio"].map(correccion)
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

df
```
